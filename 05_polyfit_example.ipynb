```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# poly fit example using numpy
```

```python
vel = np.loadtxt('v_z_25m.txt')
depths = np.arange(0,20025,25)
plt.plot(depths,vel)
plt.xlabel('depth [m]')
plt.ylabel('Velocity [m/s]')
print(vel.shape,depths.shape)
```

```python
coefs = np.polyfit(depths,vel,deg=1)
p = np.poly1d(coefs)
coef = list(coefs)
vel_fit = p(depths)

plt.plot(depths,vel)
plt.plot(depths,vel_fit)
plt.xlabel('depth [m]')
plt.ylabel('Velocity [m/s]')
```
